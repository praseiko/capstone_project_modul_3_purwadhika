### **1. Business Problem Understanding**

Pada bagian ini kita akan mencoba untuk memahami permasalahan bisnis yang terjadi, dimulai dari bisnis apa yang dijalankan, dimana bisnis tersebut dijalankan, faktor apa saja yang dapat menghambat atau mensukseskan bisnis yang dijalankan serta model seperti apa yang diperlukan dalam menunjang pengembangan bisnis dimaksud.



#### **1.1. Context**

Trasnportasi merupakan sebuah kegiatan yang berkaitan dengan perpindahan manusia atau barang dari satu tempat ke tempat lainnya dengan menggunakan sebuah kendaraan yang digerakkan oleh manusia atau mesin. Pada era saat ini alat transportasi sangat variatif, dan alat transportasi tersebut semata untuk menunjang kebutuhan manusia dalam menjalankan aktivitas kesehariannya baik untuk kepentingan bisnis maupun kepentingan pribadi. Selain itu, alat transportasi pada masa kini sudah mampu membantu kegiatan manusia baik dari jalur darat, laut ataupun udara.
Salah satu alat transportasi yang terbilang cukup populer untuk digunakan karena memiliki nilai ergonomis yang baik adalah mobil. Mobil dapat melindungi penggunanya dari cuaca panas maupun dari derasnya hujan serta lebih nyaman digunakan untuk kepentingan pribadi di banyak kondisi. Dengan nilai ergonomis yang ditawarkan, mobil menjadi kendaraan pribadi yang banyak diminati oleh seluruh kalangan.

Saudi Arabia merupakan salah satu negara penghasil minyak mentah terbesar di dunia tepatnya adalah negara kedua setelah Amerika. Menurut informasi dari [CNBC Indonesia](https://www.cnbcindonesia.com/market/20220906162442-17-369783/10-negara-penghasil-minyak-bumi-terbesar-ada-indonesia) Saudi Arabia mampu memproduksi 10,95 juta barel per hari atau berkontribusi terhadap 12,2% produksi minyak dunia pada Tahun 2021. Hal tersebut tentunya akan berpengaruh terhadap permintaan kendaraan atau alat trasnportasi dikarenakan produksi minyak yang besar akan berpengaruh juga terhadap harga jual bahan bakar minyak di negara tersebut. Maka mobil sebagai salah satu alat transportasi yang menggunakan bahan bakar minyak di arab saudi menjadi alat trasnporstasi yang digemari.

Pada saat ini bisnis jual beli mobil tidak hanya melakukan penjualan untuk mobil yang baru, namun juga untuk mobil yang sudah pernah digunakan atau mobil bekas dengan harga jual yang lebih murah. Bahkan menurut informasi dari [bangkapos.com](https://bangka.tribunnews.com/2022/03/09/segini-harga-mobil-mewah-di-arab-saudi-ternyata-lebih-murah-dibandingkan-di-indonesia?page=all) harga mobil mewah di Saudi Arabia pun relatif lebih murah jika dibandingkan dengan mobil mewah yang dijual di Indonesia. Tidak dapat dipungkiri bahwa bisnis jual beli mobil bekas menjadi salah satu kegiatan bisnis yang memiliki prospek yang menjanjikan. Hal ini juga didukung dengan peraturan negara Saudi Arabia yang memperbolehkan perempuan mengemudi pada akhir 2017 silam, serta murahnya bahan bakar di Saudi Arabia.


#### **1.2. Problem Statement**

Dari aktivitas bisnis penjulan mobil bekas, tentunya memiliki potensi peluang dan permasalahan yang dapat timbul seperti harga jual mobil bekas yang tentunya kerap kali tidak menentu. Tentu untuk menjual mobil bekas dengan harga terbaik akan menjadi pilihan yang bagus dalam menjalankan bisnis jual beli mobil bekas dengan harapan tidak menjual mobil secara overprice atau underprice yang berpotensi memiliki masalah masing-masing.
Apabila mobil dijual dengan harga yang terlalu tinggi (over price) maka berpotensi menghilangkan minat pembeli, namun sebaliknya bila mobil bekas dijual dengan harga terlalu rendah (under price) maka perusahaan berpotensi mengalami minim profit. Maka diperlukan suatu metode agar mobil yang dijual memiliki nilai harga terbaik. Penentuan harga ini tentunya tetap mengacu kepada beberapa hal seperti tipe mobil, kondisi mesin, jarak tempuh, dan sebagainya.



#### **1.3. Goals**

Berdasarkan permasalahan bisnis yang telah dijabarkan, potensi permasalahan dalam bisnis yang utama adalah berkaitan dengan harga mobil bekas yang akan dijual, karena kondisi mobil bekas memiliki harga yang beragam akibat beberapa kondisi mobil yang berbeda seperti kondisi mesin, kondisi jarak tempuh dan misalnya juga pengaruh dari transmisi mobil tersebut. Perusahaan memerlukan suatu 'tool' yang dapat memprediksi harga mobil bekas agar mendapatkan harga terbaik yang tidak *overprice* maupun *underprice*.

#### **1.4. Analytic Approach**

Yang perlu dilakukan selanjutnya adalah menganalisa data dengan mencari pola dari fitur-fitur yang ada, dengan membedakan antara mobil yang satu dengan mobil lainnya. Kemudian dari hasil analisa tersebut kita dapat membuat model regressi sebagai 'tool' prediksi yang dapat digunakan untuk memprediksi harga mobil bekas yang akan dijual.


#### **1.5. Metric Evaluation**

Evaluasi metric yang akan digunakan adalah RMSE, MAE, dan MAPE. Dimana RSME adalah nilai rataan akar kuadrat dari error, MAE adalah rataan nilai absolut dari error, sedangkan MAPE adalah rataan persentase error yang dihasilkan oleh model regresi. Semakin kecil nilai RMSE, MAE, dan MAPE yang dihasilkan, berarti model semakin akurat dalam memprediksi harga sewa sesuai dengan limitasi fitur yang digunakan. 

Selain itu, kita juga bisa menggunakan nilai R-squared atau adj. R-squared jika model yang nanti terpilih sebagai final model adalah model linear. Nilai R-squared digunakan untuk mengetahui seberapa baik model dapat merepresentasikan varians keseluruhan data. Semakin mendekati 1, maka semakin fit pula modelnya terhadap data observasi. Namun, metrik ini tidak valid untuk model non-linear.

*Outliers* merupakan data pencilan atau data yang menyimpang jauh (Ekstrim) dari data lainnya. *Outliers* dapat mempengaruhi atau membuat analisa terhadap data menjadi bias atau tidak sesuai dengan kondisi sebenarnya, terutama pada pembuatan model *Machine Learning*.

Selanjutnya kita akan mencari tahu terkait dengan *Outliers*, apakah *Outliers* yang terdapat dalam Dataset tersebut merupakan *Outliers* yang positif atau negatif dan apakah nanti penanganan terhadap *Outliers* tersebut akan berpengaruh terhadap dataset yang dianalisa.

Dalam statistik, data outliers sering menimbulkan hasil yang bias. Oleh karena itu, harus diberikan perlakuan khusus. Pengeluaran data outliers atau penggunaan data outliers tidak semata-mata merujuk kepada statistiknya, tetapi juga *adjustment* dari peneliti. Berdasarkan sumber tersebut di atas, kita tetap akan melakukan adjustment pada hasil analisa *Outlier* yang dilakukan berdasarkan kondisi nyata dengan membandingkan beberapa sumber yang relevan dengan analisis dataset untuk menghindari potensi kehilangan banyak data. Selanjutnya kita akan membuat fungsi untuk melihat *outliers* dan juga kita akan melihat *outliers* per fitur.

```python
# Outliers check function

def outliers_check(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f"IQR : {iqr}")

    batas_bawah = q1 - limit
    batas_atas = q1 + limit
    print(f"Batas Bawah :{batas_bawah}")
    print(f"Batas Atas :{batas_atas}")
```

#### **3.3. Feature Engineering**

Pada bagian ini kita akan melakukan 'Feature Egineering' dimana kegiatan ini bertujuan untuk memberikan input yang lebih baik pada model *Machine Learning*.

**Encoding**

Encoding merupakan suatu metode yang dapat diterapkan untuk merepresentasikan variabel kategorik dalam machine learning. Terdapat berbagai macam jenis metode encoding diantaranya one hot encoding, ordinal encoding, dan binary encoding.

```python
# Encoding menggunakan Ordinal Encoding untuk data Ordinal atau memiliki strata (tingkatan)
ordinal_mapping = [{
    'col' : 'Options',
    'mapping' : {None:0, 'Standard':1, 'Semi Full':2, 'Full':3}
}]

ordinal_encoder = ce.OrdinalEncoder(cols = 'Options', mapping = ordinal_mapping)
```

```python
# Transform fitur kategorikal menjadi numerikal menggunakan Column Transformer

transformer = ColumnTransformer([
    ('One Hot', OneHotEncoder(drop='first', handle_unknown='ignore'), ['Gear_Type']),
    ('Ordinal', ordinal_encoder, ['Options']),
    ('Binary', ce.BinaryEncoder(), ['Make', 'Type', 'Region'])
    ], remainder = 'passthrough')
```

```python
# Define Algoritma
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state = 42)
rf = RandomForestRegressor(random_state = 42)
xgb = XGBRegressor(random_state = 42)

log_lr = TransformedTargetRegressor(lr, func = np.log, inverse_func = np.exp)
log_knn = TransformedTargetRegressor(knn, func = np.log, inverse_func = np.exp)
log_dt = TransformedTargetRegressor(dt, func = np.log, inverse_func = np.exp)
log_rf = TransformedTargetRegressor(rf, func = np.log, inverse_func = np.exp)
log_xgb = TransformedTargetRegressor(xgb, func = np.log, inverse_func = np.exp)

# Scaling data manggunakan standar scaler
scaler = StandardScaler()

# Kandidat Algoritma
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]

score_r2 = []

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

# Mencari model terbaik berdasrakan tiap metric
for i in models:

    crossval = KFold(n_splits = 5, shuffle = True, random_state = 42)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    #RMSE
    model_cv_rmse = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv = crossval,
        scoring = 'neg_root_mean_squared_error',
        error_score = 'raise')
    
    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    #MAE
    model_cv_mae = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv = crossval,
        scoring = 'neg_mean_absolute_error',
        error_score = 'raise')
    
    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    #MAPE
    model_cv_mape = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv = crossval,
        scoring = 'neg_mean_absolute_percentage_error',
        error_score = 'raise')
    
    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())

    #R-Square
    r2_cv = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv = crossval,
        scoring = 'r2',
        error_score = 'raise')
    
    print(r2_cv, i)

    score_r2.append(r2_cv)
```

```python
# Evaluasi hasil dari 5 kandidat algoritma yang digunakan

pd.DataFrame({
    'Model' : ['Linear Regression','KNN Regressor', 'DecisionTree Regressor', 'RandomForest Regressor','XGBoost Regressor'],
    'mean_RMSE' : nilai_mean_rmse,
    'std_RMSE' : nilai_std_rmse,
    'mean_MAE' : nilai_mean_mae,
    'std_mae' : nilai_std_mae,
    'mean_MAPE' : nilai_mean_mape,
    'std_MAPE' : nilai_std_mape
})
```

MAE mengukur besarnya rata-rata kesalahan dalam serangkaian prediksi, tanpa mempertimbangkan arahnya. Ini adalah rata-rata sampel uji dari perbedaan absolut antara prediksi dan pengamatan aktual di mana semua perbedaan individu memiliki bobot yang sama. Sedangkan RMSE adalah aturan penilaian kuadrat yang juga mengukur besaran rata-rata kesalahan. Ini adalah akar kuadrat dari rata-rata perbedaan kuadrat antara prediksi dan observasi aktual. ([medium.com](https://medium.com/human-in-a-machine-world/mae-and-rmse-which-metric-is-better-e60ac3bde13d.))

Jadi, berdasarkan evaluasi kandidat algoritma yang telah kita lakukan terhadap 5 model yang diujikan, terdapat beberapa informasi yang perlu dijabarkan sebagai berikut:

Hal ini dikarenakan pada RMSE nilai residual nya atau nilai errornya dikuadratkan terlebih dahulu sebelum dirata-ratakan. Sehingga terdapat nilai-nilai error yang lebih besar oleh semua algoritma yang digunakan dan tentunya hal tersebut menjadi penyebab perbedaan nilai yang signifikan antara RMSE dengan MAE.

**Hyperparameter Tuning**

Pada tahapan ini kita akan melakukan tuning pada model yang kita nilai sebagai model terbaik untuk mendapatkan performa model yang tentunya lebih baik setelah dilakukan proses hyperparameter tuning. Model yang akan digunakan seusia dengan hasil analisa adalah XGBoost.

Acuan untuk melakukan tuning dengan model XGBoost mengacu pada [XGBoost Tuning ](https://xgboost.readthedocs.io/en/latest/parameter.html#general-parameters)

```python
# Kedalaman pohon
max_depth = list(np.arange(1,11))

# Learning rate
learning_rate = list(np.arange(1,100)/100)

# Jumlah pohon
n_estimators = list(np.arange(100,201))

# Jumlah baris tiap pohon (% dari total baris train set)
subsample = list(np.arange(2,10)/10)

# Gamma (min_impurity_decrease)
gamma = list(np.arange(1,11))

# Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = list(np.arange(1,10)/10)

# Alpha (Regularization)
reg_alpha = list(np.logspace(-3, 1, 10))

# Hyperparamater space XGBoost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=42, verbosity=0)

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=0)

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 50,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_root_mean_squared_error',
    random_state = 42
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

```python
# Melihat hasil tuning dalam bentuk dataframe. Diurutkan berdasarkan RMSE, MAE & MAPE
pd.DataFrame(random_xgb.cv_results_).sort_values(by=['rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_error', 'rank_test_neg_mean_absolute_percentage_error']).head()
```

```python
# Cek skor dan params terbaik
print('XGBoost')
print('Best_score:', random_xgb.best_score_)
print('Best_params:', random_xgb.best_params_)
```

**Predict to Test Set with The Tuned Model**

Melakukan prediksi pada set test dengan menggunakan model XGBoost dan Hyperparameter terpilih.

```python
# Model XGBoost
model = {'XGB' : XGBRegressor(random_state = 42)}

# Define Model terhadap estimator terbaik
xgb_tuning = random_xgb.best_estimator_

# Fitting Model
xgb_tuning.fit(X_train, y_train)

# Predict test set
y_pred_xgb_tuning = xgb_tuning.predict(X_test)

# Simpan nilai metrics RMSE, MAE & MAPE setelah tuning
rmse_xgb_tuning = np.sqrt(mean_squared_error(y_test, y_pred_xgb_tuning))
mae_xgb_tuning = mean_absolute_error(y_test, y_pred_xgb_tuning)
mape_xgb_tuning = mean_absolute_percentage_error(y_test, y_pred_xgb_tuning)

score_after_tuning = pd.DataFrame({'RMSE' : rmse_xgb_tuning, 'MAE' : mae_xgb_tuning, 'MAPE' : mape_xgb_tuning}, index = model.keys())
score_after_tuning
```
